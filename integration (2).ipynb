```python
import os
import mailbox
import sys
import pprint
 
print("Reading emails:")
 
mbox_file = r"C:\Users\hp\project\phishing-URL-detection-master\phishing1.mbox"
 
print("Processing " + mbox_file)
mbox = mailbox.mbox(mbox_file)
 
for key in mbox.iterkeys():
 
    try:
        message = mbox[key]
    except mbox.errors.MessageParseError:
        continue  # The message is malformed. Just leave it.
 
    #print("From: " + message['from'])
    #print("To: " +str(message['To']))
    #print ("Subject: " +str(message['Subject']))
    print("Message-ID: " +str(message['Message-ID']))
    #print(message)
 
```

```python
if message.is_multipart():
    content = ''.join(part.get_payload(decode=True) for part in message.get_payload())
else:
    content = message.get_payload(decode=True)
```
